### view HI

```python
DATASET = 'Bearing1_1'
x = np.load(f'../../../data/10FEMTO/processed_data/{DATASET}/{DATASET}_X_2560.npy')
y = np.load(f'../../../data/10FEMTO/processed_data/{DATASET}/{DATASET}_Y.npy')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.optimize import fsolve

FEMTO = '../../../data/10FEMTO/raw_data/Validation_set/'
LEN = 2560

accs = os.listdir(FEMTO+'Bearing3_3')
accs.sort()
accs = [acc for acc in accs if acc.startswith('acc')]
life = len(accs)
print(int(life * 0.82))
def equation(tau, a, T):
    return 1 + np.exp(a) - np.exp((T * tau) + a)

initial_guess = 0
a = 1
T = 15
result = fsolve(equation, initial_guess, args=(a, T))
tau = result[0]
print(f"The solution for Ï„ is: {tau}")
def HI(t):
    return 1 + np.exp(a) - np.exp((t * tau) + a)

rul = [i for i in range(life)]
hi_y = list(map(HI, rul))
min_value = min(hi_y)
max_value = max(hi_y)
normalized_values = [(x - min_value) / (max_value - min_value) for x in hi_y]

plt.figure(figsize=(8.61, 3.09))
plt.xlim(0, 4000)
plt.ylim(0.0, 1.0)
plt.plot(normalized_values, linewidth=2)
plt.axvline(x=life*0.8, color='r', linewidth=1)
plt.axhline(y=0.82, color='r', linewidth=1)
plt.show()
```

### Train Test Dataset Vibration Signal

```python
def draw_vibration(FOLDERS: List[str], title: str):
    FEMTO = '../../../data/10FEMTO/raw_data/Validation_set/'
    ax: Axes3D = plt.figure().add_subplot(111, projection='3d', facecolor='w')
    for (i, folder) in enumerate(FOLDERS):
        accs = os.listdir(FEMTO+folder)
        accs.sort()
        accs = [acc for acc in accs if acc.startswith('acc')]
        y_list = []
        x_list = []
        for stamp, acc in enumerate(accs):
            x = pd.read_csv(f'{FEMTO}/{folder}/{acc}', header=None, sep=',', usecols=[4])
            # x_avg = x.sum() / len(x)
            # x_max = x.max()
            x_median = np.median(x)
            # sorted_data = np.sort(x)
            # x_34 = np.percentile(sorted_data, 75)
            y_list.append(x_median)
            x_list.append(stamp)
        y_list = np.array(y_list)
        y_list = y_list.reshape(-1)
        
        ax.plot(xs=x_list, ys=y_list, zs=-i, zdir='x', label='curve in (x, y)', linewidth=0.3)

    ax.set_xticks(np.arange(0, -len(FOLDERS), -1), labels=FOLDERS, rotation=45, horizontalalignment='right', verticalalignment='center_baseline')
    ax.set_xlim(-len(FOLDERS) + 1, 0)

    ax.set_ylabel('Time ($ 10^4$s)')
    ax.set_yticks(np.arange(0, 4000, 1000), labels=[f'{i}' for i in range(4)])
    ax.set_ylim(0, 3000)

    ax.set_zlabel('Vibration amplitudes(g)')
    ax.set_title(title, fontsize=18)
    plt.show()
```

```python
TRAIN_DATASET = ['Bearing1_1', 'Bearing1_2', 'Bearing2_1', 'Bearing2_2', 'Bearing3_1', 'Bearing3_2']
draw_vibration(TRAIN_DATASET, 'Training data')
```

```python
TEST_DATASET = ['Bearing1_3', 'Bearing1_4', 'Bearing1_5', 'Bearing1_6', 'Bearing1_7', \
            'Bearing2_3', 'Bearing2_4', 'Bearing2_5', 'Bearing2_6', 'Bearing2_7',\
            'Bearing3_3']
draw_vibration(TEST_DATASET, 'Testing data')
```

### DL vs. TL HI Index

```python
def dl_tl_hi(np_path, TEST_DATASET, PRETRAIN, FINETUNE, y):
    hi = np.load(f'../../../data/10FEMTO/processed_data/{y}/{y}_Y.npy')
    hi_dl = np.load(f'{np_path}/{FINETUNE}_pretrain_7_{TEST_DATASET}.npy')
    hi_tl = np.load(f'{np_path}/{PRETRAIN}_pretrain_{FINETUNE}_finetune_7_{TEST_DATASET}.npy')
    plt.plot(hi_dl, linewidth=0.7, label='DL')
    plt.plot(hi_tl, linewidth=0.7, label='TL')
    plt.plot(hi, linewidth=0.7, label='Desired Response', linestyle='--')
    plt.legend()
    plt.xlabel('Time (s)')
    plt.ylabel('Health Index HI')
    plt.title(rf'{TEST_DATASET} $\rightarrow$ {PRETRAIN}')
    plt.xlim(0, len(hi))
    plt.ylim(0, 1)
    plt.show()
```

```python
TEST_DATASET = 'Bearing1_4'
PRETRAIN = 'Bearing3'
FINETUNE = 'Bearing1'
dl_tl_hi('../../../data/np/', TEST_DATASET, PRETRAIN, FINETUNE, TEST_DATASET)
```
