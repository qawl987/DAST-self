### view HI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.optimize import fsolve

FEMTO = '../../../data/10FEMTO/raw_data/Validation_set/'
LEN = 2560

accs = os.listdir(FEMTO+'Bearing3_3')
accs.sort()
accs = [acc for acc in accs if acc.startswith('acc')]
life = len(accs)
print(int(life * 0.82))
def equation(tau, a, T):
    return 1 + np.exp(a) - np.exp((T * tau) + a)

initial_guess = 0
a = 1
T = 15
result = fsolve(equation, initial_guess, args=(a, T))
tau = result[0]
print(f"The solution for Ï„ is: {tau}")
def HI(t):
    return 1 + np.exp(a) - np.exp((t * tau) + a)

rul = [i for i in range(life)]
hi_y = list(map(HI, rul))
min_value = min(hi_y)
max_value = max(hi_y)
normalized_values = [(x - min_value) / (max_value - min_value) for x in hi_y]

plt.figure(figsize=(8.61, 3.09))
plt.xlim(0, 4000)
plt.ylim(0.0, 1.0)
plt.plot(normalized_values, linewidth=2)
plt.axvline(x=life*0.8, color='r', linewidth=1)
plt.axhline(y=0.82, color='r', linewidth=1)
plt.show()
```

## vibration data

```python
import os
FEMTO = '../../../data/10FEMTO/raw_data/Validation_set/'
folder = 'Bearing1_4_norm'
accs = os.listdir(FEMTO+folder)
accs.sort()
accs = [acc for acc in accs if acc.startswith('acc')]
x_list = []
for stamp, acc in enumerate(accs):
    x = pd.read_csv(f'{FEMTO}/{folder}/{acc}', header=None, sep=',')
    x_avg = x.sum() / len(x)
    x_list.append(x_avg)
x_list = np.array(x_list)
print(x_list.shape)
plt.plot(x_list, marker='o', markersize=1)
plt.ylim(min(x_list), max(x_list))
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label', )
plt.title(f'Bearing1-1 Vibration')
# Display the plot
plt.show()
```
