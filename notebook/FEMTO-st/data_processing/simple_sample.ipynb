```python
import numpy as np
```

```python
num_samples = 2040
slope = 3  # 正比的斜率
intercept = 2  # 截距
all_sample = []
window_size = 40
feature_windows = []
y_windows = []
```

```python
for i in range(1):
    features = np.linspace(i, i+1, num_samples)
    all_sample.append(features)
```

```python
features = np.transpose(np.array(all_sample))
```

```python
output = np.array(np.linspace(0, 1, num_samples))
```

```python
features.shape
```

```python
# np.save(f'../../../data/10FEMTO/processed_data/simple_data_2040_20_X', features)
# np.save(f'../../../data/10FEMTO/processed_data/simple_data_2040_20_Y', output)
```

```python
for i in range(num_samples - window_size):
    feature_window = features[i:i + 40, :]
    y_window = output[i + 40]
    feature_windows.append(feature_window)
    y_windows.append(y_window)
```

```python
feature_windows = np.array(feature_windows)
y_windows = np.array(y_windows)
```

```python
import matplotlib.pyplot as plt

# Example data

# Plotting the line chart
plt.plot(features, label='Pred', marker='o', markersize=1)
plt.plot(output, label='Y', marker='s', markersize=1)
# Adding labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Train Loss Curve')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
